```python
import numpy as np
```

```python
x = np.linspace(-3, 3)
y = np.linspace(-3, 3)
```

```python
xv, yv = np.meshgrid(x, y)
```

```python
def rosenbrock(x, y):
    return 100 * (y - x ** 2) ** 2 + (1 - x) ** 2
```

```python
def sphere(x, y):
    return (x - np.array([1] * len(x))) ** 2 + (y - np.array([2] * len(y))) ** 2
```

```python
# z = rosenbrock(xv, yv)
z = sphere(xv, yv)
```

```python
import matplotlib.pyplot as plt
%pylab inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xv, yv, z)
```

```python
data = []
for x_, y_, z_ in zip(xv.flatten().tolist(), yv.flatten().tolist(), z.flatten().tolist()):
    data.append([x_, y_, z_])
```
